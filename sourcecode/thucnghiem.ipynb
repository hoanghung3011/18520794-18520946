```
from numpy import random
import numpy as np
```

```
def isSubsetSum(set, n, sum,count):
     
    # The value of subset[i][j] will be 
    # true if there is a
    count=0
    # subset of set[0..j-1] with sum equal to i
    subset =([[False for i in range(sum + 1)] 
            for i in range(n + 1)])
    count+=5
     
    # If sum is 0, then answer is true 
    for i in range(n + 1):
        subset[i][0] = True
        count+=1
        count+=2
    # If sum is not 0 and set is empty, 
    # then answer is false 
    for i in range(1, sum + 1):
         subset[0][i]= False
         count+=2 
         count+=1      
    # Fill the subset table in botton up manner
    for i in range(1, n + 1):
        count+=2
        for j in range(1, sum + 1):
            count+=2
            if j<set[i-1]:
                subset[i][j] = subset[i-1][j]
                count+=2
            if j>= set[i-1]:
                count+=2
                count+=2
                subset[i][j] = (subset[i-1][j] or
                                subset[i - 1][j-set[i-1]])
                count+=6
    # uncomment this code to print table 
    # for i in range(n + 1):
    # for j in range(sum + 1):
    # print (subset[i][j], end =" ")
    # print()
    # return subset[n][sum]
    return count
```

```
c=0
arrN=[]
arrC=[]
for i in range (0,50):
    S=random.randint(1000)
    val=random.randint(100,size=(random.randint(1,50)))
    n=len(val)
    arrN.append(n*S)
    arrC.append(isSubsetSum(val,n,S,c))

dic = {'n*sum': arrN , 'T(n*sum)': arrC}
```

Tính độ phức tạp



```
df_cal = df.copy()
df_cal["lg(n*sum)"] = np.log2(df_cal["n*sum"])
df_cal["sqrt(n*sum)"] = np.sqrt(df_cal["n*sum"])
df_cal["n*sum"] = df_cal["n*sum"]
df_cal["(n*sum)lg(n*sum)"] = df_cal["n*sum"] * np.log2(df_cal["n*sum"])
df_cal["(n*sum)^2"] = df_cal["n*sum"]**2
df_cal["(n*sum)^3"] = df_cal["n*sum"]**3
df_cal["2^(n*sum)"] = [2**int(i) for i in df_cal["n*sum"]]
df_cal["(n*sum)!"] = np.array([(np.math.factorial(int(i)))for i in df_cal["n*sum"]])
df_cal.to_csv("data_cal.csv")
```

```
# n*sum	T(n*sum)	lg(n*sum)	sqrt(n*sum)	(n*sum)lg(n*sum)	(n*sum)^2	(n*sum)^3
data_cop = df_cal[['n*sum',	'T(n*sum)',	'lg(n*sum)',	'sqrt(n*sum)',	'(n*sum)lg(n*sum)',	'(n*sum)^2',	'(n*sum)^3']].copy()
data_cop.to_csv('data_cropped.csv')
data_cop
```

Linear Coefficient

```
dt = pd.read_csv('/content/data.csv')
dt.columns = ['size', 'operations']
```

```
from sklearn.linear_model import LinearRegression
```

```
# lg(n*W)
X_1 = np.log2(dt['size'])
X_1 = X_1.values.reshape(-1,1)
regr = LinearRegression()
regr.fit(X_1, dt['operations'])
co2 = float(regr.coef_)
print(co2)
```

```
# sqrt(n*W)
X_2 = dt['size'].values.reshape(-1,1)
X_2 = X_2**(1/2)
regr = LinearRegression()

regr.fit(X_2, dt['operations'])
co1 = float(regr.coef_)
# print(float(regr.coef_))
print(co1)
```

```
# n*W
X_3 = dt['size'].values.reshape(-1,1)
regr = LinearRegression()
regr.fit(X_3,dt['operations'])
co3 = float(regr.coef_)
print(co3)
```

```
# (n*W)lg(n*W)
n = np.round(np.log2(dt['size']).values.reshape(-1,1))
X_4 = np.round(dt['size'].values.reshape(-1,1))
X_4 = X_4 * n
regr = LinearRegression()
regr.fit(X_4, dt['operations'])
co4 = float(regr.coef_)
print(co4)
```

```
# (n*W)^2
X_5 = dt['size'].values.reshape(-1,1)
X_5 = X_5**2
regr = LinearRegression()

regr.fit(X_5, dt['operations'])
co5 = float(regr.coef_)
print(co5)
```

```
# (n*W)^3
X_6 = dt['size'].values.reshape(-1,1)
X_6 = X_6**3
regr = LinearRegression()
regr.fit(X_6, dt['operations'])
co6 = float(regr.coef_)
print(co6)
```

Driver code

```
# n*sum	T(n*sum)	lg(n*sum)	sqrt(n*sum)	(n*sum)lg(n*sum)	(n*sum)^2	(n*sum)^3
data_coef = data_cop.copy()
data_coef["lg(n*sum)*{}".format(co1)] = data_coef["lg(n*sum)"] * co1
data_coef["sqrt(n*sum)*{}".format(co2)] = data_coef["sqrt(n*sum)"] * co2
data_coef["n*sum*{}".format(co3)] = data_coef["n*sum"] * co3
data_coef["(n*sum)lg(n*sum)*{}".format(co4)] = data_coef["(n*sum)lg(n*sum)"] * co4
data_coef["(n*sum)^2*{}".format(co5)] = data_coef["(n*sum)^2"] * co5
data_coef["(n*sum)^3*{}".format(co6)] = data_coef["(n*sum)^2"] * co6
```
